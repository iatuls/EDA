### Adding Avg Score and Total Score in our data.

```python
exams['total_score']=(exams['reading score']+exams['math score']+exams['writing score'])
exams['avg_score']=(exams['total_score']/3)
```

```python
exams.head()
```

### Time for some graphs and analysing data

```python
plt.figure(figsize=(15,6))
plt.subplot(121)
sns.histplot(data=exams,x='avg_score',bins=35,kde=True,color='#3361FF')

plt.subplot(122)
sns.histplot(data=exams,x='avg_score',bins=30,kde=True,color='#FF33EC',hue='gender')
```

```python
plt.figure(figsize=(15,6))
plt.subplot(131)
sns.boxplot(data=exams,x='math score',y='gender',hue='lunch')

plt.subplot(132)
sns.boxplot(data=exams,x='reading score',y='gender',hue='lunch')

plt.subplot(133)
sns.boxplot(data=exams,x='writing score',y='gender',hue='lunch')
```

```python
sns.barplot(x=exams['race/ethnicity'].value_counts().index,
            y=exams['race/ethnicity'].value_counts().values)

plt.xlabel('race/ethnicity')
plt.ylabel('frequency')
plt.title('race/ethinicity barplot')
plt.show()
```

```python
sns.histplot(data=exams,x='avg_score',hue='race/ethnicity',kde=True,bins=30)
```
